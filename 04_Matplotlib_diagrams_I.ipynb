```python
import numpy as np
import matplotlib.pyplot as plt

#Plotting Mathematical Functions
x_values = np.linspace(0,20,100)
y_values = np.sin(x_values)

plt.plot(x_values,y_values)
plt.show()
```

```python
x = np.linspace(0,10,100)
y = (6*x-30)**2
plt.plot(x,y)
plt.show()          #function (6x-30)²
```

```python
#Visualizing values
numbers = 10*np.random.random(100)
plt.plot(numbers,'bo')
plt.show()
```

```python
#Multiple graphs
x = np.linspace(0,5,200)
y1 = 2*x
y2 = x**2
y3 = np.log(x)

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
#Subplots       #show in the same window but independently from each other
x = np.linspace(0,5,200)
y1 = np.sin(x)
y2 = np.sqrt(x)

plt.subplot(211)        #grid of window     #rows, columns, index
plt.plot(x,y1,'r-')

plt.subplot(212)
plt.plot(x,y2,'g--')

plt.show()
```

```python
#Multiple plotting windows
plt.figure(1)
plt.plot(x, y1,'r-')

plt.figure(2)
plt.plot(x, y2, 'g--')
```

```python
#Plotting styles
from matplotlib import style 
style.use('ggplot')
```

```python
#Labeling diagrams

#setting titles

x = np.linspace(0,50,100)
y = np.sin(x)

plt.title('Sine Function')
plt.suptitle('Data Science')

plt.grid(True)      #shows the grid

plt.plot(x,y)
plt.figure("MyFigure")

plt.show()
```
